```python
import os
import datetime

import IPython
import IPython.display
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import tensorflow as tf
from statsmodels.nonparametric.smoothers_lowess import lowess


#predict step
OUT_STEPS = 3
#input step
IN_STEPS = 100
#epochs
set_epochs = 60
#use solar cycle data
solar_cycle = True
#LOESS rating
frac_rate = 0.08
solar_frac_rate = 0.05


#Load datasets
solar_cycle_csv_path = 'Solar_Cycle.csv'
monthly_temp_csv_path = 'ncei.noaa.gov_data_noaa-global-surface-temperature_v5.1_access_timeseries_aravg.mon.land_ocean.90S.90N.v5.1.0.202307.csv'
solar_cycle_df = pd.read_csv(solar_cycle_csv_path)
monthly_temp_df = pd.read_csv(monthly_temp_csv_path)

#Solar_Cycle.csv form
#time-tag: yyyy-mm
#ssn: mean monthly S.I.D.C. sunspot number
#smoothed_ssn: smoothed S.I.D.C. sunspot number
#observed_swpc_ssn: mean monthly SWPC/SWO sunspot number
#smoothed_swpc_ssn: smoothed SWPC/SWO sunspot number
#f10.7: mean monthly Penticton, B.C. 10.7cm radio flux values (sfu)
#smoothed_f10.7: smoothed 10.7cm radio flux values (sfu)
if solar_cycle:
    #cut solar_cycle to fit length of monthly_temp
    #del after 2023-07
    index_to_keep = solar_cycle_df[solar_cycle_df['time-tag'] == '2023-07'].index[0]  #      
    solar_cycle_df = solar_cycle_df.iloc[:index_to_keep + 1]
    #del before 1850-01
    index_to_keep = solar_cycle_df[solar_cycle_df['time-tag'] == '1850-01'].index[0]
    solar_cycle_df = solar_cycle_df.iloc[index_to_keep:]
    #not use
    solar_cycle_df.drop(['time-tag', 'smoothed_ssn', 'observed_swpc_ssn', 'smoothed_swpc_ssn', 'smoothed_f10.7', 'f10.7'], axis=1, inplace=True)
    #reset index
    solar_cycle_df = solar_cycle_df.reset_index(drop=True)
    
monthly_temp_df.drop(['total-error-variance', 'high-frequency-error-variance', 'low-frequency-error-variance', 'bias-error-variance', 'diagnostic variable', 'diagnostic variable.1', 'diagnostic variable.2'], axis=1, inplace=True)

if solar_cycle:
    #combine dataset
    df = pd.concat([monthly_temp_df, solar_cycle_df], axis=1)
else:
    df = monthly_temp_df

#time data
date_time = pd.to_datetime(monthly_temp_df.pop('year'), format='%Y')
date_time = pd.to_datetime(monthly_temp_df.pop('month'), format='%m')

#LOEWSS
y = df['temp'].dropna()
x = np.arange(0,len(y))
res = lowess(endog=y,exog=x,frac=frac_rate,it=3)
df['lowess_temp'] = res[:,1]
df.drop(['temp'], axis=1, inplace=True)


if solar_cycle:
    y = df['ssn'].dropna()
    res = lowess(endog=y,exog=x,frac=solar_frac_rate,it=3)
    df['lowess_ssn'] = res[:,1]
    df.drop(['ssn'], axis=1, inplace=True)


#fit to LSTM
def Sequential_Input_LSTM(df, input_sequence):
    df_np = df.to_numpy()
    X = []
    y = []
    
    for i in range(len(df_np) - input_sequence):
        row = [a for a in df_np[i:i + input_sequence]]
        X.append(row)
        label = df_np[i + input_sequence]
        y.append(label)
        
    return np.array(X), np.array(y)

    

df_min_model_data = df['lowess_temp']

X, y = Sequential_Input_LSTM(df_min_model_data, IN_STEPS)

#partitioning data
n = len(X)
# Training data(70%)
X_train, y_train = X[0:int(n*0.7)], y[0:int(n*0.7)]

# Validation data(20%)
X_val, y_val = X[int(n*0.7):int(n*0.9)], y[int(n*0.7):int(n*0.9)]

# Test data(10%)
X_test, y_test = X[int(n*0.9):], y[int(n*0.9):]

from tensorflow.keras.models import Sequential, save_model, load_model
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam 

#number of feature
n_features = 1                        
#use Sequential
model1 = Sequential()
#layer form
model1.add(InputLayer((IN_STEPS,n_features)))
model1.add(LSTM(100, return_sequences = True))     
model1.add(LSTM(100, return_sequences = True))
model1.add(LSTM(50))
model1.add(Dense(8, activation = 'relu'))
model1.add(Dense(1, activation = 'linear'))
#show layer
model1.summary()
#earlystop
early_stop = EarlyStopping(monitor = 'root_mean_squared_error', patience = 2)
#learn
model1.compile(loss = MeanSquaredError(), 
               optimizer = Adam(learning_rate = 0.0001), 
               metrics = RootMeanSquaredError())

model1.fit(X_train, y_train, 
           validation_data = (X_val, y_val), 
           epochs = set_epochs, 
           callbacks = [early_stop])

#show history
losses_df1 = pd.DataFrame(model1.history.history)
losses_df1.plot(figsize = (10,6))

# save model
model1.save('LSTM_Models/lstm_univariate.keras')
# load model
model1 = load_model('LSTM_Models/lstm_univariate.keras')
```

```python
test_predictions1 = model1.predict(X_test).flatten()

X_test_list = []
for i in range(len(X_test)):
    X_test_list.append(X_test[i][0])
    
#output
test_predictions_df1 = pd.DataFrame({'X_test':list(X_test_list), 
                                    'LSTM Prediction':list(test_predictions1)})
```
